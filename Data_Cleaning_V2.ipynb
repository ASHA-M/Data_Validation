```python
# Validation check for RREL1 - values should be alphanumeric and not contain ND1, ND2, ND3, ND4, ND5 and unique
exclude_values = ["ND1", "ND2", "ND3", "ND4", "ND5"]

def first_col(value):
    if value.isalnum() and (value not in exclude_values):
        return value
    else:
        return "validation failed"

valid_data = pd.DataFrame(data['RREL1'].apply(first_col))
```

```python
#Storing validation checks in this dataframe
valid_data.head(1000) 
```

```python
#Checking if columns to append GBP are numeric

def is_column_numeric(column):
    return pd.to_numeric(column, errors="coerce").notna().all()

# List of monetary columns to check
columns_to_check = ["RREL16", "RREL20", "RREL29", "RREL30", "RREL31", "RREL32", "RREL33", "RREL39", "RREL41", "RREL61", "RREL64", "RREL67", "RREL71", 
                     "RREL73", "RREL74", "RREL77", "RREC13", "RREC21"]

for column in columns_to_check:
    if is_column_numeric(data[column]):
        print(f"{column} contains all numeric values.")
    else:
        print(f"{column} does not contain all numeric values.")
# Some columns are not numeric these non-numeric columns contain ND1-ND5
```

```python
# Append GBP to the following monetary columns if numeric and apply validation checker:

columns_to_modify = ["RREL16", "RREL20", "RREL29", "RREL30", "RREL31", "RREL32", "RREL33", "RREL39", "RREL41", "RREL61", "RREL64", "RREL67", "RREL71", 
                     "RREL73", "RREL74", "RREL77", "RREC13", "RREC21"]

def nd14checker(value): #Checks for ND1-ND4 and returns Valid if allowed else returns value - ND1-ND4:YES, ND5:NO
    if value in ['ND1', 'ND2', 'ND3', 'ND4']:
        return 'Valid'
    else:
        return value

def nd15checker(value): #Checks for ND1-ND5 returns Valid else returns value - ND1-ND4: YES ND5: YES
    if value in ['ND1', 'ND2', 'ND3', 'ND4', 'ND5']:
        return 'Valid'
    else:
        return value
    
def nd5checker(value): #Checks for ND5 only returns Valid else returns value - ND1-ND4: NO NDS: YES
    if value in ['ND5']:
        return 'Valid'
    else:
        return value

    
for column in columns_to_modify: #Checks for 3 conditions here for each using functions above.
                                 #Checks the data type of each value in the column and conditionally add "GBP" to numeric values.
    if column in ['RREL20', 'RREL29','RREL31' 'RREL32', 'RREL41','RREL64', 'RREC13']: # nd15 checker both conditions are YES
        valid_data[column] = data[column].apply(lambda x: str(x) + " GBP" if pd.api.types.is_number(x) else nd15checker(x))
    elif column in ['RREL30', 'RREL33', 'RREL39', 'RREL61', 'RREL71', 'RREL73', 'RREL74', 'RREL77', 'RREC21']: # nd5checker only ND5 is YES
        valid_data[column] = data[column].apply(lambda x: str(x) + " GBP" if pd.api.types.is_number(x) else nd5checker(x))
    elif column in ['RREL16']: # nd14 only ND1-ND4 is YES
        valid_data[column] = data[column].apply(lambda x: str(x) + " GBP" if pd.api.types.is_number(x) else nd14checker(x))
    else:
        valid_data[column] = data[column].apply(lambda x: str(x) + " GBP" if pd.api.types.is_number(x) else x)

valid_data.head(1000)
```

```python
# Convert date format to YYYY-MM-DD (keep only datepart not the timestamp) and apply validator checker: 

columns_to_convert = ["RREL6", "RREL7", "RREL8", "RREL9", "RREL23", "RREL24", "RREL36", "RREL51", "RREL53", "RREL55", "RREL60", "RREL62", "RREL63",
                      "RREL65", "RREL66", "RREL72", "RREC15", "RREC19", "RREC20"]
    
#Function converting to datetime and keeping only the datepart
def convert_to_datetime(cell_value):
    try:
        datetime_obj = pd.to_datetime(cell_value)
        date_only = datetime_obj.date()
        return date_only
    except ValueError:
        return 'error'

for column in columns_to_convert: # Convert specified columns to datetime or leave as is if not a valid date.
                                  # Checks for 3 conditions here for each using functions used previously. 
    if column in ['RREL23', 'RREC19']: # nd14 only ND1-ND4 is YES
        valid_data[column] = data[column].apply(lambda x: convert_to_datetime(x) if convert_to_datetime(x) != 'error' else nd14checker(x))
    elif column in ['RREL7','RREL8','RREL9', 'RREL24', 'RREL36', 'RREL51', 'RREL53', 'RREL55', 'RREL72', ]: # nd5checker only ND5 is YES
        valid_data[column] = data[column].apply(lambda x: convert_to_datetime(x) if convert_to_datetime(x) != 'error' else nd5checker(x))
    elif column in ['RREL60','RREL62', 'RREL63', 'RREL65', 'RREL66', 'RREC15', 'RREC20']:  # nd15 checker both conditions are YES
        valid_data[column] = data[column].apply(lambda x: convert_to_datetime(x) if convert_to_datetime(x) != 'error' else nd15checker(x))
    else:
        valid_data[column] = data[column].apply(convert_to_datetime)

valid_data.head(1000)
```

```python
valid_data['Index'] = list(range(1,valid_data.shape[0] + 1))
```

```python
valid_data.set_index('Index', inplace = True)

ordered_columns = ['RREL1', 'RREL6', 'RREL7', 'RREL8', 'RREL9', 'RREL16', 'RREL20', 'RREL23', 'RREL24', 'RREL29', 'RREL30', 'RREL31', 'RREL32','RREL33', 
                   'RREL36', 'RREL39', 'RREL41', 'RREL51', 'RREL53', 'RREL55', 'RREL60', 'RREL61', 'RREL62', 'RREL63', 'RREL64', 'RREL65', 'RREL66',
                   'RREL67', 'RREL71', 'RREL72', 'RREL73', 'RREL74', 'RREL77', 'RREC13', 'RREC15', 'RREC19', 'RREC20', 'RREC21']

valid_data.columns = ordered_columns
```

```python
#To preserve date format converting to .xlsx file instead of .csv
with pd.ExcelWriter('cleaned_data_V2.xlsx', engine='xlsxwriter') as writer:
    valid_data.to_excel(writer, sheet_name='Sheet1', index=True, startrow=0, header=True)
```
